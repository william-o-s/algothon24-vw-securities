```python
# Import required libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Import data
df = pd.read_csv('prices.txt', sep='   ', header=None, names=[f"stock{i}" for i in range(50)])
df = df.T
df.head()
```

```python
# Plot all stocks
plt.figure(num='Universe', figsize=(20,20))
for index, row in df.iterrows():
    plt.plot(row, label=index)
plt.legend()
plt.show()
```

```python
# Plot specific stock
stock = 'stock0'

plt.figure(num=stock, figsize=(5,5))
plt.plot(df.loc[stock], label=stock)
```

```python
# Plot all stocks individually
rows = 5
cols = 10

fig, axs = plt.subplots(nrows=rows, ncols=cols, figsize=(20,10))
fig.tight_layout()

for i in range(rows):
    for j in range(cols):
        index = (i * cols) + j
        axs[i, j].plot(df.iloc[index])
        axs[i, j].set_title(df.iloc[index].name)
plt.show()
```
